# Análise Exploratória de Dados - Imóveis de São Paulo

## Propósito

Este notebook realiza a análise exploratória dos dados imobiliários da cidade de São Paulo, especificamente dados de IPTU (Imposto Predial e Territorial Urbano) e ITBI (Imposto sobre Transmissão de Bens Imóveis) para o período de 2019 a 2024. O objetivo é compreender a estrutura dos dados, identificar padrões, valores ausentes e características das variáveis numéricas e categóricas.

## Entradas (Inputs)

### Dados de IPTU
- **Arquivo CSV**: `data/IPTU_2019_2024.csv`
  - Dataset processado contendo informações de IPTU com Valor Venal calculado

### Dados de ITBI
- **Arquivo CSV**: `data/ITBI_2019_2024.csv`
  - Dataset consolidado de transações imobiliárias

1. **Carregamento de Dados**: Importa os datasets de IPTU e ITBI
2. **Análise de Valores Ausentes**: Identifica e quantifica dados faltantes

```python
# Importa biblioteca pandas para manipulação de dados
import pandas as pd

# Configura formatação de números float para exibição com 2 casas decimais
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

## Carregamento e Visualização dos Dados de IPTU

Carrega o dataset de IPTU e exibe as primeiras linhas e informações sobre a estrutura dos dados.


```python
# Carrega o dataset de IPTU processado
df_iptu = pd.read_csv('data/IPTU_2019_2024.csv')

# Exibe as primeiras 20 linhas do dataset
display(df_iptu.head(20))

# Exibe informações sobre a estrutura do dataset (tipos de dados, valores não nulos, uso de memória)
display(df_iptu.info())
```

### Conversão de Tipos de Dados

Converte colunas identificadoras para tipo object (string) para preservar zeros à esquerda e facilitar manipulações.


```python
# Converte número do contribuinte para string para preservar zeros à esquerda
df_iptu["NUMERO DO CONTRIBUINTE"] = df_iptu["NUMERO DO CONTRIBUINTE"].astype("object")

# Converte ano do exercício para string
df_iptu["ANO DO EXERCICIO"] = df_iptu["ANO DO EXERCICIO"].astype("object")
```

## Análise de Valores Ausentes

Identifica e quantifica valores ausentes (NaN) em cada coluna do dataset de IPTU.

```python
# Calcula a quantidade de valores ausentes por coluna
missing_values = df_iptu.isnull().sum()
display(missing_values)
```

## Gera Estatísticas Descritivas

Calcula estatísticas descritivas (média, desvio padrão, quartis, mínimo, máximo) para todas as variáveis numéricas do dataset de IPTU.


```python
# Seleciona apenas as colunas numéricas (int64 e float64)
numerical_columns = df_iptu.select_dtypes(include=['int64', 'float64'])

# Calcula estatísticas descritivas (count, mean, std, min, 25%, 50%, 75%, max)
descriptive_stats = numerical_columns.describe()
display(descriptive_stats)
```

## Analisa Dados Categóricos

Examina a distribuição e frequências das variáveis categóricas do dataset de IPTU. Para colunas com muitos valores únicos, exibe apenas as 20 categorias mais frequentes.


```python
# Seleciona todas as colunas do tipo object (categóricas)
categorical_columns = df_iptu.select_dtypes(include='object').columns.tolist()

# Remove colunas identificadoras que não são variáveis categóricas de interesse
categorical_columns.remove('NUMERO DO CONTRIBUINTE')
categorical_columns.remove('ANO DO EXERCICIO')

# Para cada coluna categórica, exibe a contagem de valores
for col in categorical_columns:
    print(f"\nContagem de valores para coluna: {col}")
    # Exibe as 20 categorias mais frequentes para colunas com muitos valores únicos
    if df_iptu[col].nunique() > 20:
        display(df_iptu[col].value_counts().head(20))
    else:
        # Exibe todas as categorias se houver 20 ou menos valores únicos
        display(df_iptu[col].value_counts())
```

## Carregamento e Visualização dos Dados de ITBI

Carrega o dataset de ITBI e exibe as primeiras linhas e informações sobre a estrutura dos dados.


```python
# Carrega o dataset de ITBI processado
# low_memory=False evita avisos sobre tipos mistos durante a leitura
df_itbi = pd.read_csv('data/ITBI_2019_2024.csv', low_memory=False)

# Exibe as primeiras 20 linhas do dataset
display(df_itbi.head(20))

# Exibe informações sobre a estrutura do dataset (tipos de dados, valores não nulos, uso de memória)
display(df_itbi.info())
```

### Análise de Valores Ausentes - ITBI

Identifica e quantifica valores ausentes (NaN) em cada coluna do dataset de ITBI.


```python
# Calcula a quantidade de valores ausentes por coluna no dataset de ITBI
missing_values = df_itbi.isnull().sum()
display(missing_values)
```

### Estatísticas Descritivas - ITBI

Calcula estatísticas descritivas para todas as variáveis numéricas do dataset de ITBI.


```python
# Seleciona apenas as colunas numéricas (int64 e float64) do dataset de ITBI
numerical_columns = df_itbi.select_dtypes(include=['int64', 'float64'])

# Calcula estatísticas descritivas (count, mean, std, min, 25%, 50%, 75%, max)
descriptive_stats = numerical_columns.describe()
display(descriptive_stats)
```

### Análise de Dados Categóricos - ITBI

Examina a distribuição e frequências das variáveis categóricas do dataset de ITBI. Para colunas com muitos valores únicos, exibe apenas as 20 categorias mais frequentes.


```python
# Seleciona todas as colunas do tipo object (categóricas) do dataset de ITBI
categorical_columns = df_itbi.select_dtypes(include='object').columns.tolist()
display(categorical_columns)

# Remove coluna identificadora que não é variável categórica de interesse
categorical_columns.remove('N° do Cadastro (SQL)')

# Para cada coluna categórica, exibe a contagem de valores
for col in categorical_columns:
    print(f"\nContagem de valores para coluna: {col}")
    # Exibe as 20 categorias mais frequentes para colunas com muitos valores únicos
    if df_itbi[col].nunique() > 20:
        display(df_itbi[col].value_counts().head(20))
    else:
        # Exibe todas as categorias se houver 20 ou menos valores únicos
        display(df_itbi[col].value_counts())
```
